# Pequeño Reto
En el último vídeo de la unidad sobre "Agrupar las filas de un dataFrame", que explica cómo usar apply sobre datos agrupados, se indica que, siempre que sea posible, es recomendable usar funciones directamente sobre los datos agrupados, en lugar de usar apply, por ser más eficiente.

Escribid un pequeño script, donde calculais, sobre el dataframe "notas",  para cada expediente, la nota ponderada sin usar bucles ni apply, sino directamente manipulaciones y cálculos sobre columnas

```python
from pathlib import Path

import pandas as pd

DATA_DIRECTORY: Path = Path("../data")

notas: pd.DataFrame = pd.read_csv(DATA_DIRECTORY / "notas.csv")
notas
```

```python
# Primero se obtiene la nota multiplicada por los ects que vale para cada fila

notas["ponderada"] = notas["ects"] * notas["nota"]

# Esto hará una columna nueva que usaremos con los grupos
notas
```

```python
# Ahora las agrupo por expediente y calculo la media ponderada
grouped = notas.groupby("expediente")

suma = grouped["ponderada"].sum()

ects = grouped["ects"].sum()

expedientes_ponderados = suma / ects

expedientes_ponderados
```
